## üìù Exerc√≠cio 1: Organizando Lista de Produtos
**Cen√°rio**: Voc√™ tem uma lista de produtos com pre√ßos. Precisa organizar e filtrar os dados.

**Tarefas**:
1. Remover produtos duplicados
2. Filtrar produtos com pre√ßo acima de R$ 50
3. Ordenar por pre√ßo (menor para maior)
4. Contar quantos produtos restaram

**Dica**: Use opera√ß√µes b√°sicas de listas e loops!

```python
# Lista de produtos (alguns duplicados)
produtos = [
    ("Mouse", 25.90),
    ("Teclado", 89.00),
    ("Monitor", 299.99),
    ("Mouse", 25.90),  # duplicado
    ("Fone", 45.50),
    ("Webcam", 120.00),
    ("Teclado", 89.00),  # duplicado
    ("Mousepad", 15.99),
    ("Caixa de Som", 75.00)
]

print("� Lista original de produtos:")

for produto, preco in produtos:
    print(f"  {produto}: R$ {preco:.2f}")

print(f"\nüìä Total de itens na lista original: {len(produtos)}")

# ‚úèÔ∏è TAREFA 1: Remover duplicados
# Dica: Voc√™ pode converter para set e depois para list, ou usar um loop
produtos_unicos = []
# Seu c√≥digo aqui:
seen = set()
for produto in produtos:
    if produto not in seen:
        seen.add(produto)
        produtos_unicos.append(produto)

print("\nLista de produtos √∫nicos:")
for produto, preco in produtos_unicos:
    print(f"  {produto}: R$ {preco:.2f}")

print(f"\nüìä Total de itens ap√≥s remover duplicados: {len(produtos_unicos)}")



# ‚úèÔ∏è TAREFA 2: Filtrar produtos com pre√ßo > R$ 50
produtos_caros = []
# Seu c√≥digo aqui:
for produto, preco in produtos_unicos:
    if preco > 50:
        produtos_caros.append((produto, preco))

print("\nLista de produtos caros (acima de R$ 50):")
for produto, preco in produtos_caros:
    print(f"  {produto}: R$ {preco:.2f}")


# ‚úèÔ∏è TAREFA 3: Ordenar por pre√ßo (menor para maior)
# Dica: Use sorted() com key=lambda
produtos_ordenados = []
# Seu c√≥digo aqui:
produtos_ordenados = sorted(produtos_unicos, key=lambda x: x[1])

print("\nLista de produtos ordenados por pre√ßo (menor para maior):")
for produto, preco in produtos_ordenados:
    print(f"  {produto}: R$ {preco:.2f}")


# ‚úèÔ∏è TAREFA 4: Mostrar resultado final
print("\nüéØ RESULTADO FINAL:")
print("Produtos √∫nicos com pre√ßo > R$ 50, ordenados por pre√ßo:")
for produto, preco in produtos_caros:
    print(f"  {produto}: R$ {preco:.2f}")

print(f"\nüìä Total de produtos finais: {len(produtos_ordenados)}")
```

## üë• Exerc√≠cio 2: Cadastro de Clientes com Dicion√°rios
**Cen√°rio**: Organize informa√ß√µes de clientes usando dicion√°rios para facilitar buscas e atualiza√ß√µes.

**Tarefas**:
1. Filtrar clientes por cidade (S√£o Paulo)
2. Encontrar clientes ativos (mais de 3 compras)
3. Calcular a idade m√©dia dos clientes
4. Contar quantos clientes h√° por cidade

**Dica**: Use loops e opera√ß√µes com dicion√°rios para acessar e filtrar os dados!

```python
# Base de dados de clientes
clientes = {
    "12345": {"nome": "Ana Silva", "idade": 28, "cidade": "S√£o Paulo", "compras": 3},
    "67890": {"nome": "Jo√£o Santos", "idade": 35, "cidade": "Rio de Janeiro", "compras": 7},
    "11111": {"nome": "Maria Costa", "idade": 42, "cidade": "S√£o Paulo", "compras": 2},
    "22222": {"nome": "Pedro Lima", "idade": 29, "cidade": "Belo Horizonte", "compras": 5},
    "33333": {"nome": "Julia Ferreira", "idade": 33, "cidade": "S√£o Paulo", "compras": 1}
}

print("üë• Base de clientes atual:")
for id_cliente, dados in clientes.items():
    print(f"  ID {id_cliente}: {dados['nome']} - {dados['cidade']} - {dados['compras']} compras")

# ‚úèÔ∏è TAREFA 1: Encontrar clientes de S√£o Paulo
print("\nÔøΩ TAREFA 1: Clientes de S√£o Paulo")
clientes_sp = []
# Seu c√≥digo aqui:
for id_cliente, dados in clientes.items():
    if dados["cidade"] == "S√£o Paulo":
        clientes_sp.append(dados["nome"])

print("Clientes de S√£o Paulo:", clientes_sp)


# ‚úèÔ∏è TAREFA 2: Encontrar clientes com mais de 3 compras
print("\nüîç TAREFA 2: Clientes com mais de 3 compras")
clientes_ativos = []
# Seu c√≥digo aqui:
for id_clientes, dados in clientes.items():
    if dados["compras"] >= 3:
        clientes_ativos.append(dados["nome"])

print("Clientes com 3 compras ou mais:", clientes_ativos)




# ‚úèÔ∏è TAREFA 3: Calcular idade m√©dia dos clientes
print("\nüìä TAREFA 3: Idade m√©dia dos clientes")
# Seu c√≥digo aqui:
idades = [dados["idade"] for dados in clientes.values()]
idade_media = sum(idades) / len(idades)
print(f"Idade m√©dia dos clientes: {idade_media:.0f} anos")

# ‚úèÔ∏è TAREFA 4: Contar clientes por cidade
print("\nüèôÔ∏è TAREFA 4: Clientes por cidade")
contagem_cidades = {}
# Seu c√≥digo aqui:
for id_cliente, dados in clientes.items():
    cidade = dados["cidade"]
    contagem_cidades[cidade] = contagem_cidades.get(cidade, 0) + 1

for cidade, quantidade in contagem_cidades.items():
    print(f"{cidade}: {quantidade} {'clientes' if quantidade > 1 else 'cliente'}")
```

## üéØ Exerc√≠cio 3: Opera√ß√µes com Conjuntos (Sets)
**Cen√°rio**: Analise interesses de usu√°rios usando opera√ß√µes de conjuntos.

**Tarefas**: Descobrir interesses comuns, √∫nicos e diferen√ßas entre grupos.

```python
# Interesses de diferentes grupos de usu√°rios
grupo_A = {"futebol", "m√∫sica", "filmes", "games", "culin√°ria", "viagem"}
grupo_B = {"m√∫sica", "leitura", "filmes", "arte", "fotografia", "viagem"}
grupo_C = {"games", "tecnologia", "m√∫sica", "programa√ß√£o", "filmes"}

print("üéØ Interesses dos grupos:")
print(f"Grupo A: {grupo_A}")
print(f"Grupo B: {grupo_B}")
print(f"Grupo C: {grupo_C}")

# ‚úèÔ∏è TAREFA 1: Encontrar interesses comuns a todos os grupos (interse√ß√£o)
print("\nü§ù TAREFA 1: Interesses comuns a TODOS os grupos")
interesses_comuns = set()
# Seu c√≥digo aqui (use intersection ou &):
interesses_comuns = grupo_A & grupo_B & grupo_C
print(interesses_comuns)


# ‚úèÔ∏è TAREFA 2: Encontrar todos os interesses √∫nicos (uni√£o)
print("\nüåü TAREFA 2: TODOS os interesses √∫nicos")
todos_interesses = set()
# Seu c√≥digo aqui (use union ou |):
todos_interesses = grupo_A | grupo_B | grupo_C
print(todos_interesses)  # ou grupo_A.union(grupo_B, grupo_C)


# ‚úèÔ∏è TAREFA 3: Interesses exclusivos do Grupo A (diferen√ßa)
print("\nÔøΩ TAREFA 3: Interesses EXCLUSIVOS do Grupo A")
exclusivos_A = set()
# Seu c√≥digo aqui (use difference ou -):
exclusivos_A = grupo_A - grupo_B - grupo_C
print(exclusivos_A)  # ou grupo_A.difference(grupo_B, grupo_C)

# ‚úèÔ∏è TAREFA 4: Interesses comuns entre A e B, mas n√£o em C
print("\n‚ö° TAREFA 4: Comuns entre A e B, mas N√ÉO em C")
comuns_AB_nao_C = set()
# Seu c√≥digo aqui:
comuns_AB_nao_C = (grupo_A & grupo_B) - grupo_C
print(comuns_AB_nao_C)
```

## üìä Exerc√≠cio 4: Agrupando Dados com Listas
**Cen√°rio**: Voc√™ tem vendas de diferentes produtos e precisa agrupar e calcular totais.

**Tarefas**: Agrupar vendas por produto e calcular estat√≠sticas b√°sicas.

```python
# Lista de vendas: (produto, quantidade, valor_unitario)
vendas = [
    ("Notebook", 2, 2500.00),
    ("Mouse", 10, 25.50),
    ("Teclado", 5, 85.00),
    ("Notebook", 1, 2500.00),
    ("Monitor", 3, 450.00),
    ("Mouse", 15, 25.50),
    ("Teclado", 2, 85.00),
    ("Webcam", 8, 120.00),
    ("Monitor", 1, 450.00)
]

print("üí∞ Lista de vendas:")
for produto, qtd, valor in vendas:
    total = qtd * valor
    print(f"  {produto}: {qtd} unidades √ó R$ {valor:.2f} = R$ {total:.2f}")

# ‚úèÔ∏è TAREFA 1: Calcular total de vendas por produto
print("\nüìä TAREFA 1: Total de vendas por produto")
vendas_por_produto = {}
# Seu c√≥digo aqui:
# Dica: Para cada venda, some quantidade √ó valor_unitario no dicion√°rio
for produto, qtd, valor in vendas:
    total = qtd * valor
    if produto in vendas_por_produto:
        vendas_por_produto[produto] += total
    else:
        vendas_por_produto[produto] = total

for produto, total in vendas_por_produto.items():
    print(f"{produto}: R$ {total:.2f}")

# ‚úèÔ∏è TAREFA 2: Calcular quantidade total vendida por produto
print("\nüì¶ TAREFA 2: Quantidade total por produto")
quantidade_por_produto = {}
# Seu c√≥digo aqui:
for produto, qtd, _ in vendas:
    if produto in quantidade_por_produto:
        quantidade_por_produto[produto] += qtd
    else:
        quantidade_por_produto[produto] = qtd

for produto, qtd in quantidade_por_produto.items():
    print(f"{produto}: {qtd} unidades")

# ‚úèÔ∏è TAREFA 3: Encontrar o produto mais vendido (em quantidade)
print("\nüèÜ TAREFA 3: Produto mais vendido")
produto_mais_vendido = ""
maior_quantidade = 0
# Seu c√≥digo aqui:
for produto, qtd in quantidade_por_produto.items():
    if qtd > maior_quantidade:
        maior_quantidade = qtd
        produto_mais_vendido = produto

print(f"O produto mais vendido foi '{produto_mais_vendido}' com {maior_quantidade} unidades.")



# ‚úèÔ∏è TAREFA 4: Calcular faturamento total
print("\nüí∏ TAREFA 4: Faturamento total")
faturamento_total = 0
# Seu c√≥digo aqui:
faturamento_total = sum(vendas_por_produto.values())
print(f"Faturamento total: R$ {faturamento_total:.2f}")
```

## üßπ Exerc√≠cio 5: Limpeza de Dados de Texto
**Cen√°rio**: Voc√™ tem uma lista de nomes de usu√°rios que precisa padronizar.

**Tarefas**: Limpar e formatar nomes usando opera√ß√µes de string.

```python
# Lista de nomes com problemas de formata√ß√£o
nomes_sujos = [
    "  ANA silva  ",
    "jo√£o PEREIRA",
    "  MARIA costa   ",
    "Pedro   Lima",
    "JULIA   ferreira  ",
    "carlos SANTOS",
    "  Beatriz  OLIVEIRA  "
]

print("üìù Nomes originais (com problemas):")
for i, nome in enumerate(nomes_sujos, 1):
    print(f"  {i}. '{nome}'")

# ‚úèÔ∏è TAREFA 1: Limpar espa√ßos em branco e padronizar capitaliza√ß√£o
print("\n‚ú® TAREFA 1: Nomes limpos e padronizados")
nomes_limpos = []
# Seu c√≥digo aqui:
# Dica: Use strip() para remover espa√ßos e title() para capitalizar
for nome in nomes_sujos:
    nome_limpo = nome.strip().title()
    nomes_limpos.append(nome_limpo)
    print(f"  {nome_limpo}")


# ‚úèÔ∏è TAREFA 2: Separar nomes e sobrenomes
print("\nüë• TAREFA 2: Separar nomes e sobrenomes")
nomes_separados = []
# Seu c√≥digo aqui:
# Dica: Use split() e crie tuplas (nome, sobrenome)

for nome_completo in nomes_limpos:
    partes = nome_completo.split()
    nome = partes[0]
    sobrenome = " ".join(partes[1:])
    nomes_separados.append((nome, sobrenome))

for nome, sobrenome in nomes_separados:
    print(f"Nome: {nome}, Sobrenome: {sobrenome}")


# ‚úèÔ∏è TAREFA 3: Criar lista de primeiros nomes √∫nicos
print("\nüéØ TAREFA 3: Primeiros nomes √∫nicos")
primeiros_nomes = set()
# Seu c√≥digo aqui:
primeiros_nomes = {nome for nome, _ in nomes_separados}
print(f"  - {primeiros_nomes}")


# ‚úèÔ∏è TAREFA 4: Contar quantos nomes t√™m mais de uma palavra
print("\nÔøΩ TAREFA 4: Nomes compostos")
nomes_compostos = 0
# Seu c√≥digo aqui:
nomes_compostos = sum(1 for nome in nomes_limpos if len(nome.split()) > 1)
print(f"  - Quantidade de nomes compostos: {nomes_compostos}")
```
