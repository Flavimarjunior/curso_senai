## ðŸ“ ExercÃ­cio 1: Organizando Lista de Produtos
**CenÃ¡rio**: VocÃª tem uma lista de produtos com preÃ§os. Precisa organizar e filtrar os dados.

**Tarefas**:
1. Remover produtos duplicados
2. Filtrar produtos com preÃ§o acima de R$ 50
3. Ordenar por preÃ§o (menor para maior)
4. Contar quantos produtos restaram

**Dica**: Use operaÃ§Ãµes bÃ¡sicas de listas e loops!

```python
# Lista de produtos (alguns duplicados)
produtos = [
    ("Mouse", 25.90),
    ("Teclado", 89.00),
    ("Monitor", 299.99),
    ("Mouse", 25.90),  # duplicado
    ("Fone", 45.50),
    ("Webcam", 120.00),
    ("Teclado", 89.00),  # duplicado
    ("Mousepad", 15.99),
    ("Caixa de Som", 75.00)
]

print("� Lista original de produtos:")

for produto, preco in produtos:
    print(f"  {produto}: R$ {preco:.2f}")

print(f"\nðŸ“Š Total de itens na lista original: {len(produtos)}")

# âœï¸ TAREFA 1: Remover duplicados
# Dica: VocÃª pode converter para set e depois para list, ou usar um loop
produtos_unicos = []
# Seu cÃ³digo aqui:
seen = set()
for produto in produtos:
    if produto not in seen:
        seen.add(produto)
        produtos_unicos.append(produto)

print("\nLista de produtos Ãºnicos:")
for produto, preco in produtos_unicos:
    print(f"  {produto}: R$ {preco:.2f}")

print(f"\nðŸ“Š Total de itens apÃ³s remover duplicados: {len(produtos_unicos)}")



# âœï¸ TAREFA 2: Filtrar produtos com preÃ§o > R$ 50
produtos_caros = []
# Seu cÃ³digo aqui:
for produto, preco in produtos_unicos:
    if preco > 50:
        produtos_caros.append((produto, preco))

print("\nLista de produtos caros (acima de R$ 50):")
for produto, preco in produtos_caros:
    print(f"  {produto}: R$ {preco:.2f}")


# âœï¸ TAREFA 3: Ordenar por preÃ§o (menor para maior)
# Dica: Use sorted() com key=lambda
produtos_ordenados = []
# Seu cÃ³digo aqui:
produtos_ordenados = sorted(produtos_unicos, key=lambda x: x[1])

print("\nLista de produtos ordenados por preÃ§o (menor para maior):")
for produto, preco in produtos_ordenados:
    print(f"  {produto}: R$ {preco:.2f}")


# âœï¸ TAREFA 4: Mostrar resultado final
print("\nðŸŽ¯ RESULTADO FINAL:")
print("Produtos Ãºnicos com preÃ§o > R$ 50, ordenados por preÃ§o:")
for produto, preco in produtos_caros:
    print(f"  {produto}: R$ {preco:.2f}")

print(f"\nðŸ“Š Total de produtos finais: {len(produtos_ordenados)}")
```

## ðŸ‘¥ ExercÃ­cio 2: Cadastro de Clientes com DicionÃ¡rios
**CenÃ¡rio**: Organize informaÃ§Ãµes de clientes usando dicionÃ¡rios para facilitar buscas e atualizaÃ§Ãµes.

**Tarefas**:
1. Filtrar clientes por cidade (SÃ£o Paulo)
2. Encontrar clientes ativos (mais de 3 compras)
3. Calcular a idade mÃ©dia dos clientes
4. Contar quantos clientes hÃ¡ por cidade

**Dica**: Use loops e operaÃ§Ãµes com dicionÃ¡rios para acessar e filtrar os dados!

```python
# Base de dados de clientes
clientes = {
    "12345": {"nome": "Ana Silva", "idade": 28, "cidade": "SÃ£o Paulo", "compras": 3},
    "67890": {"nome": "JoÃ£o Santos", "idade": 35, "cidade": "Rio de Janeiro", "compras": 7},
    "11111": {"nome": "Maria Costa", "idade": 42, "cidade": "SÃ£o Paulo", "compras": 2},
    "22222": {"nome": "Pedro Lima", "idade": 29, "cidade": "Belo Horizonte", "compras": 5},
    "33333": {"nome": "Julia Ferreira", "idade": 33, "cidade": "SÃ£o Paulo", "compras": 1}
}

print("ðŸ‘¥ Base de clientes atual:")
for id_cliente, dados in clientes.items():
    print(f"  ID {id_cliente}: {dados['nome']} - {dados['cidade']} - {dados['compras']} compras")

# âœï¸ TAREFA 1: Encontrar clientes de SÃ£o Paulo
print("\nï¿½ TAREFA 1: Clientes de SÃ£o Paulo")
clientes_sp = []
# Seu cÃ³digo aqui:
for id_cliente, dados in clientes.items():
    if dados["cidade"] == "SÃ£o Paulo":
        clientes_sp.append(dados["nome"])

print("Clientes de SÃ£o Paulo:", clientes_sp)


# âœï¸ TAREFA 2: Encontrar clientes com mais de 3 compras
print("\nðŸ” TAREFA 2: Clientes com mais de 3 compras")
clientes_ativos = []
# Seu cÃ³digo aqui:
for id_clientes, dados in clientes.items():
    if dados["compras"] >= 3:
        clientes_ativos.append(dados["nome"])

print("Clientes com 3 compras ou mais:", clientes_ativos)




# âœï¸ TAREFA 3: Calcular idade mÃ©dia dos clientes
print("\nðŸ“Š TAREFA 3: Idade mÃ©dia dos clientes")
# Seu cÃ³digo aqui:
idades = [dados["idade"] for dados in clientes.values()]
idade_media = sum(idades) / len(idades)
print(f"Idade mÃ©dia dos clientes: {idade_media:.0f} anos")

# âœï¸ TAREFA 4: Contar clientes por cidade
print("\nðŸ™ï¸ TAREFA 4: Clientes por cidade")
contagem_cidades = {}
# Seu cÃ³digo aqui:
for id_cliente, dados in clientes.items():
    cidade = dados["cidade"]
    contagem_cidades[cidade] = contagem_cidades.get(cidade, 0) + 1

for cidade, quantidade in contagem_cidades.items():
    print(f"{cidade}: {quantidade} {'clientes' if quantidade > 1 else 'cliente'}")
```

## ðŸŽ¯ ExercÃ­cio 3: OperaÃ§Ãµes com Conjuntos (Sets)
**CenÃ¡rio**: Analise interesses de usuÃ¡rios usando operaÃ§Ãµes de conjuntos.

**Tarefas**: Descobrir interesses comuns, Ãºnicos e diferenÃ§as entre grupos.

```python
# Interesses de diferentes grupos de usuÃ¡rios
grupo_A = {"futebol", "mÃºsica", "filmes", "games", "culinÃ¡ria", "viagem"}
grupo_B = {"mÃºsica", "leitura", "filmes", "arte", "fotografia", "viagem"}
grupo_C = {"games", "tecnologia", "mÃºsica", "programaÃ§Ã£o", "filmes"}

print("ðŸŽ¯ Interesses dos grupos:")
print(f"Grupo A: {grupo_A}")
print(f"Grupo B: {grupo_B}")
print(f"Grupo C: {grupo_C}")

# âœï¸ TAREFA 1: Encontrar interesses comuns a todos os grupos (interseÃ§Ã£o)
print("\nðŸ¤ TAREFA 1: Interesses comuns a TODOS os grupos")
interesses_comuns = set()
# Seu cÃ³digo aqui (use intersection ou &):
interesses_comuns = grupo_A & grupo_B & grupo_C
print(interesses_comuns)


# âœï¸ TAREFA 2: Encontrar todos os interesses Ãºnicos (uniÃ£o)
print("\nðŸŒŸ TAREFA 2: TODOS os interesses Ãºnicos")
todos_interesses = set()
# Seu cÃ³digo aqui (use union ou |):
todos_interesses = grupo_A | grupo_B | grupo_C
print(todos_interesses)  # ou grupo_A.union(grupo_B, grupo_C)


# âœï¸ TAREFA 3: Interesses exclusivos do Grupo A (diferenÃ§a)
print("\nï¿½ TAREFA 3: Interesses EXCLUSIVOS do Grupo A")
exclusivos_A = set()
# Seu cÃ³digo aqui (use difference ou -):
exclusivos_A = grupo_A - grupo_B - grupo_C
print(exclusivos_A)  # ou grupo_A.difference(grupo_B, grupo_C)

# âœï¸ TAREFA 4: Interesses comuns entre A e B, mas nÃ£o em C
print("\nâš¡ TAREFA 4: Comuns entre A e B, mas NÃƒO em C")
comuns_AB_nao_C = set()
# Seu cÃ³digo aqui:
comuns_AB_nao_C = (grupo_A & grupo_B) - grupo_C
print(comuns_AB_nao_C)
```

## ðŸ“Š ExercÃ­cio 4: Agrupando Dados com Listas
**CenÃ¡rio**: VocÃª tem vendas de diferentes produtos e precisa agrupar e calcular totais.

**Tarefas**: Agrupar vendas por produto e calcular estatÃ­sticas bÃ¡sicas.

```python
# Lista de vendas: (produto, quantidade, valor_unitario)
vendas = [
    ("Notebook", 2, 2500.00),
    ("Mouse", 10, 25.50),
    ("Teclado", 5, 85.00),
    ("Notebook", 1, 2500.00),
    ("Monitor", 3, 450.00),
    ("Mouse", 15, 25.50),
    ("Teclado", 2, 85.00),
    ("Webcam", 8, 120.00),
    ("Monitor", 1, 450.00)
]

print("ðŸ’° Lista de vendas:")
for produto, qtd, valor in vendas:
    total = qtd * valor
    print(f"  {produto}: {qtd} unidades Ã— R$ {valor:.2f} = R$ {total:.2f}")

# âœï¸ TAREFA 1: Calcular total de vendas por produto
print("\nðŸ“Š TAREFA 1: Total de vendas por produto")
vendas_por_produto = {}
# Seu cÃ³digo aqui:
# Dica: Para cada venda, some quantidade Ã— valor_unitario no dicionÃ¡rio
for produto, qtd, valor in vendas:
    total = qtd * valor
    if produto in vendas_por_produto:
        vendas_por_produto[produto] += total
    else:
        vendas_por_produto[produto] = total

for produto, total in vendas_por_produto.items():
    print(f"{produto}: R$ {total:.2f}")

# âœï¸ TAREFA 2: Calcular quantidade total vendida por produto
print("\nðŸ“¦ TAREFA 2: Quantidade total por produto")
quantidade_por_produto = {}
# Seu cÃ³digo aqui:
for produto, qtd, _ in vendas:
    if produto in quantidade_por_produto:
        quantidade_por_produto[produto] += qtd
    else:
        quantidade_por_produto[produto] = qtd

for produto, qtd in quantidade_por_produto.items():
    print(f"{produto}: {qtd} unidades")

# âœï¸ TAREFA 3: Encontrar o produto mais vendido (em quantidade)
print("\nðŸ† TAREFA 3: Produto mais vendido")
produto_mais_vendido = ""
maior_quantidade = 0
# Seu cÃ³digo aqui:
for produto, qtd in quantidade_por_produto.items():
    if qtd > maior_quantidade:
        maior_quantidade = qtd
        produto_mais_vendido = produto

print(f"O produto mais vendido foi '{produto_mais_vendido}' com {maior_quantidade} unidades.")



# âœï¸ TAREFA 4: Calcular faturamento total
print("\nðŸ’¸ TAREFA 4: Faturamento total")
faturamento_total = 0
# Seu cÃ³digo aqui:
faturamento_total = sum(vendas_por_produto.values())
print(f"Faturamento total: R$ {faturamento_total:.2f}")
```

## ðŸ§¹ ExercÃ­cio 5: Limpeza de Dados de Texto
**CenÃ¡rio**: VocÃª tem uma lista de nomes de usuÃ¡rios que precisa padronizar.

**Tarefas**: Limpar e formatar nomes usando operaÃ§Ãµes de string.

```python
# Lista de nomes com problemas de formataÃ§Ã£o
nomes_sujos = [
    "  ANA silva  ",
    "joÃ£o PEREIRA",
    "  MARIA costa   ",
    "Pedro   Lima",
    "JULIA   ferreira  ",
    "carlos SANTOS",
    "  Beatriz  OLIVEIRA  "
]

print("ðŸ“ Nomes originais (com problemas):")
for i, nome in enumerate(nomes_sujos, 1):
    print(f"  {i}. '{nome}'")

# âœï¸ TAREFA 1: Limpar espaÃ§os em branco e padronizar capitalizaÃ§Ã£o
print("\nâœ¨ TAREFA 1: Nomes limpos e padronizados")
nomes_limpos = []
# Seu cÃ³digo aqui:
# Dica: Use strip() para remover espaÃ§os e title() para capitalizar
for nome in nomes_sujos:
    nome_limpo = nome.strip().title()
    nomes_limpos.append(nome_limpo)
    print(f"  {nome_limpo}")


# âœï¸ TAREFA 2: Separar nomes e sobrenomes
print("\nðŸ‘¥ TAREFA 2: Separar nomes e sobrenomes")
nomes_separados = []
# Seu cÃ³digo aqui:
# Dica: Use split() e crie tuplas (nome, sobrenome)

for nome_completo in nomes_limpos:
    partes = nome_completo.split()
    nome = partes[0]
    sobrenome = " ".join(partes[1:])
    nomes_separados.append((nome, sobrenome))

for nome, sobrenome in nomes_separados:
    print(f"Nome: {nome}, Sobrenome: {sobrenome}")


# âœï¸ TAREFA 3: Criar lista de primeiros nomes Ãºnicos
print("\nðŸŽ¯ TAREFA 3: Primeiros nomes Ãºnicos")
primeiros_nomes = set()
# Seu cÃ³digo aqui:
primeiros_nomes = {nome for nome, _ in nomes_separados}
print(f"  - {primeiros_nomes}")


# âœï¸ TAREFA 4: Contar quantos nomes tÃªm mais de uma palavra
print("\nï¿½ TAREFA 4: Nomes compostos")
nomes_compostos = 0
# Seu cÃ³digo aqui:
nomes_compostos = sum(1 for nome in nomes_limpos if len(nome.split()) > 1)
print(f"  - Quantidade de nomes compostos: {nomes_compostos}")
```
